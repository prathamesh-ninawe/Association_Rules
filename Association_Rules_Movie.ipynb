```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
movies = pd.read_csv('my_movies.csv')
```

```python
movies
```

```python
movies.drop(['V1', 'V2', 'V3', 'V4', 'V5'], axis=1, inplace=True)
```

```python
movies.head()
```

### Association rules with 20% Support and 60% confidence

  

```python
m = apriori(movies, min_support=0.2, use_colnames=True)
```

```python
rules = association_rules(m, metric="lift", min_threshold=0.6)
```

```python
rules.sort_values('lift', ascending=False)
rules[rules.lift>1]
```

```python
rules[['support','confidence','lift']].hist()
```

```python
plt.scatter('support','confidence',data=rules)
```

```python
sns.scatterplot(x='support',y='confidence',data=rules, hue='antecedents')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```
